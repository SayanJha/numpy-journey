```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array(
    [
        [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
        [2, 120000, 140000, 160000, 190000],  # Beijing Bites
        [3, 200000, 230000, 260000, 300000],  # Pizza Hub
        [4, 180000, 210000, 240000, 270000],  # Burger Point
        [5, 160000, 185000, 205000, 230000],  # Chai Point
    ]
)
# *---------2023-----2024-----2025---2026
```

```python
print("=== Zomato sales analysis ===")
print("\nSales data shape: ", sales_data.shape)
print("\nSample data for first three restaurants: \n", sales_data[0:3])
```

```python
# Total sales per year
sales_per_year = np.sum(sales_data, axis=0)
print(sales_per_year[1:])

# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)

# Average sales per year
avg_sales = np.mean(sales_data[:, 1:], axis=0)
print(avg_sales)

# Relative sale
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("\n", cumsum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales accross all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()


print(type(sales_data))
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: \n", vector1 + vector2)
print("\nVector Multiplication: ", vector1 * vector2)
print("\nDot Product: ", np.dot(vector1, vector2))
angle = np.arccos(
    np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
)
print(angle)
```
