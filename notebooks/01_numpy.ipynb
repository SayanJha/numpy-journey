```python
py_list = [1, 2, 3]
print("Python list multiplication ", py_list * 2)

np_array = np.array([1, 2, 3])
print("Python array multiplication ", np_array * 2)
```

```python
start = time.time()
py_list = [x * 2 for x in range(1, 100000)]
print("\nList operation time: ", time.time() - start)

start = time.time()
np_array = np.arange(1, 100000) * 2
print("Array operation time: ", time.time() - start)
```

```python
### Array reshaping

arr = np.arange(12)
print("Original Array ", arr)

reshaped = arr.reshape((3, 4))
print("\nReshaped array \n", reshaped)

flattened = reshaped.flatten()
print("\nFlattened Array \n", flattened)


# ravel (returns view instead of copy)
raveled = reshaped.ravel()
print("\nRaveled Array \n", raveled)

# Transpose

transpose = reshaped.T
print("\nTransposed Array \n", transpose)
```

```python
arr = np.arange(12)
reshaped = arr.reshape(3, 4)
print(reshaped)

"""
#* Rule
total elements must stay the same
3 * 4 = 12
"""
print(arr.reshape(4,-1))

# -1 means NumPy calculates automatically
```

## Flatten vs Ravel

```python
flat = reshaped.flatten()
print(flat)
print(reshaped)
raveled = reshaped.ravel()
print(raveled)
print(reshaped)
raveled[0] = 100
print(reshaped) # Because they share same memory
print(flat.base)
print(raveled.base)
```
