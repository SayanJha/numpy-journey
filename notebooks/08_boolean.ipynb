#### Boolean Indexing

```python
import numpy as np
```

```python
arr = np.array([10, 25, 3, 47, 8, 19])

mask = arr > 15
print(mask) # True/False

print(arr[mask]) # [25 47 19]
```

```python
print(arr[arr > 15])
```

```python
arr =  np.array([10, 25, 3, 47, 8, 19])

# and -> &
print(arr[(arr > 5) & (arr < 20)])

# or -> |
print(arr[(arr < 5) | (arr > 40)])

# not -> ~
print(arr[~(arr > 15)])
# never and, or, not. Python's keywords don't work element-wise on arrays.
```

```python
# np.where(condition, value_if_true, value_if_false)

arr = np.array([1, -2, 3, -4, 5])

result = np.where(arr > 0, arr, 0)
print(result)
```

```python
def relu(x):
    return np.where(x > 0, x, 0)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr + 10)
print(arr * np.array([1, 2, 3]))
print((arr * np.array([1, 2, 3])).shape)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

col_mean = arr.mean(axis=0)  
print(col_mean.shape)
centered = arr - col_mean  
print(centered)
print(centered.shape)

row_mean = arr.mean(axis=1, keepdims=True)
print(row_mean)
centered = arr - row_mean
print(centered)
```
