#### Linear Algebra + Stacking & Splitting

````python
import numpy as  np
````

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.dot(a, b))

arr1 = np.array([[1, 2], [4, 5]])
arr2 = np.array([[7, 8], [10, 11]])
print(arr1 @ arr2)

# (m, n) @ (n, p) -> (m, p)

"""
(3, 4) @ (4, 5) → (3, 5)  ✅
(3, 4) @ (3, 5) → ERROR   ❌  4 ≠ 3
```

---

### Why this is neural networks

A single neural network layer is literally:
```
output = activation(inputs @ weights + bias)
"""
````

````python
inputs = np.random.randn(5, 3) # 5 samples, 3 features
weights = np.random.randn(3, 4) # 3 inputs 4 -> neurons
bies = np.random.randn(4) # one bies per neuron

raw = inputs @ weights + bies # shape -> (5, 4)
output = np.where(raw > 0, raw, 0) # ReLU
print(raw)
print(output)
````

````python
arr = np.array([[1, 2], [3, 4]], dtype=float)

print(np.linalg.norm(arr))
print(np.linalg.det(arr))
print(np.linalg.inv(arr))
print(arr @ (np.linalg.inv(arr)))
print(np.linalg.eig(arr))
````

````python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

print(np.vstack([arr1, arr2])) # (4, 2)
print(np.hstack([arr1, arr2])) # (2, 4)

print(np.concatenate([arr1, arr2], axis=0)) # (4, 2)
print(np.concatenate([arr1, arr2], axis=1)) # (2, 4)
````

````python
arr = np.arange(1, 25).reshape(4, 6)

print(np.split(arr, 2, axis=0))
print(np.hsplit(arr, 3))
print(np.vsplit(arr, 4))
````

````python
np.random.seed(0)

inputs = np.random.randn(6, 4)
w1 = np.random.randn(4, 5)
b1 = np.random.randn(5)
w2 = np.random.randn(5, 3)
b2 = np.random.randn(3)

raw1 = inputs @ w1 + b1
print(raw1.shape)
output1 = np.where(raw1 > 0, raw1, 0)
print(output1.shape)

raw2 = output1 @ w2 + b2
print(raw2)
print(raw2.shape)

output0 = raw2[:, 0]
best = np.argmax(output0) # return index
print(best)


l2 = output1 / np.linalg.norm(output1, axis=1, keepdims= True)
print(l2)

stacked = np.vstack([inputs, inputs])

forward_pass1 = stacked @ w1 + b1
layer_1_output = np.where(forward_pass1 >0, forward_pass1, 0)
forward_pass = layer_1_output @ w2 + b2
print(forward_pass)
print(forward_pass.shape)

# (6, 4) @ (4, 5) = (6, 5)
"""
#? Lets take a small example
|-       -|     |-      -|   |-                                      -|
| 1  2  3 |     | 7  8   |   | 1*7 + 2*9 + 3*11      1*8 + 2*10 + 3*12|
|         | X   | 9  10  | = |                                        |
| 4  5  6 |     | 11 12  |   | 4*7 + 5*9 + 6*11      4*8 + 5*10 + 6*12|
|-       -|     |-      -|   |-                                      -|
           (2, 3)         (3, 2)                                      (2, 2)
"""
# Output layer activations should be close to linear, whereas hidden layer activations can be much more non-linear. ReLU is ideally suited for non-linear activations.
#! I had to use Quora for this

""""
inputs = np.random.randn(6, 4)
#? w1 = np.random.randn(5, 4)
b1 = np.random.randn(5)
w2 = np.random.randn(5, 3)
b2 = np.random.randn(3)

raw1 = inputs @ w1 + b1 #! ValueError
# Matrix multiplication requires the number of columns in the first matrix to equal the number of rows in the second
print(raw1.shape)
output1 = np.where(raw1 > 0, raw1, 0)
print(output1.shape)"""
````
