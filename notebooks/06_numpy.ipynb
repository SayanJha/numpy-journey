#####   Python lists are flexible — they can hold integers, strings, objects, even other lists, all mixed together. That flexibility has a cost: Python stores each element as a separate object in memory, scattered across RAM. To do math on a list, Python has to visit each object one by one.

#####   NumPy arrays (ndarray) make a deal: all elements must be the same type. In exchange, NumPy packs them into a single continuous block of memory — like a row of lockers side by side. Your CPU can then blast through that block at full speed using optimized C code, sometimes thousands of times faster than a Python loop.

```python
arr = np.random.randint(2, 19, (3, 3))
print(arr[arr % 2 == 0])

arr1 = np.arange(2, 19, 2)
reshaped_arr1 = arr1.reshape(3, 3)
print("Reshaped Array: \n", reshaped_arr1)

print("Shape", reshaped_arr1.shape)
print("Dtype: ", reshaped_arr1.dtype)
row_sum = reshaped_arr1.sum(axis=1)
print("Row sums", row_sum)
```

```python
arr = np.arange(9).reshape(3, 3)

view = arr[0:2]
view[0, 0] = 999  # Slicing gives a view same memory
print(arr)

print("")

copy = arr[0:2].copy()
copy[0, 0] = 0
print(arr)
```

```python
arr = np.arange(12)

print(arr.reshape(3, 4))
print("")
print(arr.reshape(3, -1))  # -1 is wildcard
print(arr.reshape(3, 4).flatten())
print(arr.reshape(3, 4).T)
```

```python
arr = np.arange(1, 26).reshape(5, 5)
print(arr)

print("3x3 sub-matrix from bottom-right", arr[-3:, -3:])
print("3rd column: ", arr[:, 2])
print("Reversed:", arr[::-1])

view = arr[0:3]
print("View: \n", view)
copied_array = arr.copy()
view[0:2] = 100
print("Slice is a view:\n", arr)
print("Original:\n", copied_array)
```

```python
print("Shape change")
print(arr.shape)
arr = arr.reshape(1, -1)
print(arr.shape)
```

```python
arr = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
flattened = arr.flatten()
print("Original array: ", arr)
print("Flatten: ", flattened)

flattened[2] = 10000

print("Original array: ", arr)
print("Flatten: ", flattened)

raveled = arr.ravel()
print("Original array: ", arr)
print("Raveled: ", raveled)

raveled[2] = 1000  # ravel (returns view instead of copy)

print("Original array: ", arr)
print("Raveled: ", raveled)
```

```python
arr = np.arange(1, 25).reshape(4, 6)
print(arr)
print(arr.shape)
transposed = arr.T
print(transposed.shape)

arr = arr.reshape(24, -1)
print(arr.shape)
arr = arr.reshape(-1, 24)
print(arr.shape)
arr = arr.reshape(4, -1)
print(arr.shape)
arr = arr.reshape(-1, 4)
print(arr.shape)
arr = arr.reshape(3, -1)
print(arr.shape)
arr = arr.reshape(-1, 3)
```
